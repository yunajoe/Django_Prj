```python
import time
import requests
from bs4 import BeautifulSoup as bs
from selenium import webdriver
```

```python
# C:\Windows 위치에 크롬 넣기 
# 크롬은 자동으로 업데이트되기 때문에 잘 실행되던 Selenium 코드에서 갑자기 오류가 날 수 있습니다.
# 따라서 크롬을 새로 다운로드해줘야된당

path = "C:\\Driver\\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.implicitly_wait(3)
driver.get("https://www.lifequotes.co.kr/")
time.sleep(1)

# print(driver.window_handles) 팝업창 확인.. 팝업창이 하나 있으면 리스트 요소가 2개 나옴(하나는 domain창 , 다른하나는 팝업)

# pop-up창은.. 아닌데 암튼 창 하나 close
tab = driver.find_element_by_xpath('//*[@id="sgpb-popup-dialog-main-div-wrapper"]/div/img')
tab.click()
time.sleep(1)
# 인생태그
driver.find_element_by_css_selector("span.avia-menu-text").click()
time.sleep(3)

last_height = driver.execute_script("return document.body.scrollHeight")


# 페이지 스크롤 
while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(0.5)

    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
    
data = driver.page_source



```

# BS로 페이지 크롤링  

# for문으로 5페이지까지 크롤링

```python


# url = "https://www.lifequotes.co.kr/tag/인생/page" 

for page in range(1,5):
    contents = requests.get("https://www.lifequotes.co.kr/tag/인생/page/" +str(page))
    soup = bs(contents.text)
    contents = soup.select("article.slide-entry")
    for content in contents:
        words = content.select_one("a")["title"]
        pics = content.select_one("a img")["data-src"]
        print(words, pics)     
    print(page,"page 크롤링 완료")
    print("")
       
    
```

# while문으로다가 page에 글이 없을때까지 크롤링

```python
 
page = 1
while True:    
    contents = requests.get("https://www.lifequotes.co.kr/tag/인생/page/" +str(page))
    soup = bs(contents.text)
    contents = soup.select("article.slide-entry")
    for content in contents:
        words = content.select_one("a")["title"]
        pics = content.select_one("a img")["data-src"]
        print(words, pics)     
    print(page,"page 크롤링 완료")
    print("")
    signal = soup.select("div h1.main-title.entry-title a")
    s = len([s.text for s in signal])
    if s ==0:
        break
    page = page + 1
    
```
